```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('SINASC_RO_2019.csv')
```

```python
df['GESTACAO']
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
plt.rc('figure', figsize=(10, 8)) #ajustando o tamanho
fig, axes = plt.subplots(2, 3) #abrindo um novo figure 2x3

#gerando vários countplots
sns.countplot(ax = axes[0, 0], x='APGAR1', data=df)
sns.countplot(ax = axes[0, 1], x='APGAR5', data=df)
sns.countplot(ax = axes[1, 0], x='GESTACAO', data=df).tick_params(axis="x", rotation = 90)
sns.countplot(ax = axes[1, 1], x='ESTCIVMAE', data=df)
sns.countplot(ax = axes[0, 2], x='SEXO', data=df)

#ajustar o espaçamento
plt.subplots_adjust(wspace=0.5, hspace=0.3)
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
sns.barplot(x='CONSULTAS', y='APGAR5', data=df)
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
fig, axes = plt.subplots(2, 2) #abrindo um novo figure 2x2

#gerando vários countplots
sns.countplot(ax = axes[0, 0], x='SEXO', data=df)
sns.countplot(ax = axes[0, 1], x='ESTCIVMAE', data=df)
sns.countplot(ax = axes[1, 1], x='CONSULTAS', data=df)
sns.countplot(ax = axes[1, 0], x='GESTACAO', data=df).tick_params(axis="x", rotation = 80)

#ajustar o espaçamento
plt.subplots_adjust(wspace=0.3, hspace=0.3)
```
